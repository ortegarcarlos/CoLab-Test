```python
# ---
# Title: Effectiveness Factor vs. Thiele Modulus
# Author: Carlos Ortega

# Description:
#   This notebook plots the effectiveness factor (η) as a function of the
#   Thiele modulus (φ) for spherical catalyst particles.

# References:
# Davis, Mark E. / Davis, Robert J.
# Fundamentals of Chemical Reaction Engineering (2003), p200
# McGraw-Hill Higher Education
# ---

import numpy as np
import matplotlib.pyplot as plt

# Thiele modulus range
phi = np.linspace(start=0.01, stop=20, num=100)

# Effectiveness factor for first-order reaction in a sphere:
# η = (3/φ) * (1/np.tanh(φ) - 1/φ)
eta = (3/phi) * (1/np.tanh(phi) - 1/phi)

# Plot
fig, ax = plt.subplots(figsize=(6, 5), tight_layout=True)

ax.plot(phi, eta)
# Title and axis lables
ax.set_title('Effectiveness factor versus Thiele modulus.\nIsothermal first-order reaction in a sphere.', fontsize=10)
ax.set_xlabel('Thiele Modulus ($\\phi$)', fontsize=10)
ax.set_ylabel('Effectiveness Factor ($\\eta$)', fontsize=10)

# Set logarithmic scales
ax.set_xscale('log') # Logarithmic scale for x-axis
ax.set_yscale('log') # Logarithmic scale for y-axis


ax.set_xlim([0.01, 100])
ax.set_ylim([0.1, 1.1])
```
